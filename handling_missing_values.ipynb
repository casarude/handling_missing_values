# **The fillna(method='bfill') and fillna(method='ffill') methods in Pandas.**

fillna(method='bfill')

* What it does: This method propagates the next valid observation backward to fill the missing value. In simpler terms, it uses the value that comes after the missing value to fill the gap.
* How it works: bfill stands for "backward fill." It iterates through the DataFrame (or Series) and, when it encounters a missing value (NaN), it looks for the next valid (non-NaN) value in the same column. It then copies that next valid value into the missing value's place.
* Effect: This method is useful for filling missing values when you assume that the subsequent value is a reasonable approximation for the missing data. It can be helpful for time series data or situations where data is expected to be relatively continuous.

fillna(method='ffill')

* What it does: This method propagates the last valid observation forward to fill the missing value. In other words, it uses the value that comes before the missing value to fill the gap.
* How it works: ffill stands for "forward fill." It iterates through the DataFrame (or Series) and, when it encounters a missing value (NaN), it looks for the previous valid (non-NaN) value in the same column. It then copies that previous valid value into the missing value's place.
* Effect: This method is useful for filling missing values when you assume that the preceding value is a reasonable approximation for the missing data. It's often used for filling gaps in time series data or situations where data is expected to be relatively stable over time.

Important considerations:

Direction: bfill fills from the next valid value (backward), while ffill fills from the previous valid value (forward).
Data continuity: These methods assume some degree of continuity or stability in the data. If your data is highly volatile or unpredictable, using bfill or ffill might not be appropriate.
Potential bias: Be cautious when using these methods, as they can introduce bias into your data if the missing values are not randomly distributed.

# **More advanced methods to handle missing values**

  1.K-Nearest Neighbors (KNN) Imputation:
  Concept: This method leverages the idea that similar data points tend to have similar values. It identifies 'k' nearest neighbors to a data point with missing values and uses the average (or weighted average) of their values to fill the gap.**It expects numerical data to perform these calculations.**


  Advantages: Can handle both numerical and categorical data, often provides more accurate imputations than simple methods.

  Disadvantages: KNNImputer cannot directly handle string or categorical data.

  Considerations: Computationally expensive for large datasets, sensitive to the choice of 'k' and distance metric.

  While imputing a single column might be sufficient in some cases, remember that KNN imputation considers the relationships between features. Imputing multiple related columns together can potentially lead to more accurate results.

````python
import kagglehub
import pandas as pd
from sklearn.impute import KNNImputer

# Download latest version
path = kagglehub.dataset_download("maxhorowitz/nflplaybyplay2009to2016")

df = pd.read_csv(path + "/NFL Play by Play 2009-2017 (v4).csv")

# 1. Create a KNNImputer instance
imputer = KNNImputer(n_neighbors=3)  # You can adjust the number of neighbors

# 2. Reshape the 'down' column to a 2D array (required by KNNImputer)
down_values = df['down'].values.reshape(-1, 1)

# 3. Apply the imputer to the 'down' column
imputed_down = imputer.fit_transform(down_values)  # Execution took 15 minutes!

# 4. Replace the original 'down' column with the imputed values
df['down'] = imputed_down
````

````python
columns_data = df['down'].isnull().sum()

print(f"number of null values after applying KNNImputer: {columns_data}")

print(df['down'])
````

In this scenario, I have applied KNNImputer on a single column. This is how it works:

1. Limited Information: Since KNNImputer relies on distances between data points to identify neighbors, when applied to only one column, it has limited information to work with. It essentially treats each value in that column as a single-dimensional data point.
2. Distances in 1D: The distances between data points are calculated based solely on the values in the specified column. In your case, the distances between rows would be the absolute difference between their 'down' values.
3. Neighbor Selection: The KNNImputer still tries to find the 5 nearest neighbors (based on these 1D distances) for each missing value in the 'down' column.
Imputation: The missing 'down' value is then imputed using the average of the 'down' values from those 5 nearest neighbors.

In essence, applying KNNImputer to a single column performs a localized imputation based on the values in that column alone. While it can be a valid approach in some cases, it's crucial to be aware of its limitations and consider the broader context of your dataset.
